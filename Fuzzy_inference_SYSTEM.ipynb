```python
import numpy as np
from skfuzzy import control as ctrl
import skfuzzy as fuzzy
```

```python
Quality = ctrl.Antecedent(np.arange(0,11,1),'quality')
Service = ctrl.Antecedent(np.arange(0,11,1),'service')
tip = ctrl.Consequent(np.arange(0,25,1),'tip')
```

```python
Quality.automf(3)
Service.automf(3)
```

```python
tip['low']=fuzzy.trimf(tip.universe,(0,0,13))
tip['medium']=fuzzy.trimf(tip.universe,(0,13,25))
tip['high']=fuzzy.trimf(tip.universe,(13,25,25))
```

```python
Quality['average'].view()
```

```python
Quality['good'].view()
```

```python
Service['poor'].view()
```

```python
tip.view()
```

```python
Rule1=ctrl.Rule(Quality['poor']&Service['poor'],tip['low'])
Rule2=ctrl.Rule(Quality['good']&Service['good'],tip['high'])
Rule3=ctrl.Rule(Quality['average']&Service['average'],tip['medium'])
Rule4=ctrl.Rule(Quality['poor']&Service['average'],tip['low'])
Rule5=ctrl.Rule(Quality['good']&Service['average'],tip['medium'])
```

```python
Rule1.view()
```

```python
Rule2.view()
```

```python
Rule3.view()
```

```python
Rule4.view()
```

```python
Rule5.view()
```

```python
Tipping_ctrl=ctrl.ControlSystem([Rule5])
Tipping=ctrl.ControlSystemSimulation(Tipping_ctrl)
```

```python
Tipping.input['quality']=9.5
Tipping.input['service']=5.5
```

```python
Tipping.compute()
```

```python
print(Tipping.output['tip'])
```

```python
tip.view(sim = Tipping)
```
