```python
import matplotlib.pyplot as plt
import numpy as np 
year=[2001,2002,2003,2004,2005,2006,2007,2008,2020,2022,2023]
uempl=[30,20,40,30,20,50,30,25,60,45,70]
plt.plot(year,uempl,marker='o',linestyle='-.')
plt.plot(year,uempl,marker='o')
plt.xlabel('year')
plt.ylabel('unemployement rate')
plt.title('year vs unempl')
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=["science","commerce","art"] 
#h=[200,300,350] 
w=[0.4,0.5,0.6] 
color=["red","blue","green"] 
plt.bar(x,300,w,color="red") 
plt.bar(x,100,w,color="blue") 
plt.bar(x,50,w,color="green") 
plt.xlabel('courses',color="green") 
plt.ylabel('students enrolled')
plt.title('Admission Data',color="blue")
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
year=[2001,2002,2003,2004,2005,2006,2007,2008,2020,2022,2023] 
uempl=[30,20,40,30,20,50,30,25,60,45,70] 
#supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.plot(year,uempl,marker='x',linestyle="dotted") 
plt.plot(year,uempl,marker='+') 
plt.xlabel('year', color='blue') 
plt.ylabel('unemployement rate',color='blue') 
plt.title('year vs unempl',color='c')
plt.show()
```
